```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
bank_marketing = fetch_ucirepo(id=222) 
  
# data (as pandas dataframes) 
X = bank_marketing.data.features 
y = bank_marketing.data.targets 
  
```

```python
myrows=['age','job','marital','education','default','balance','housing','loan','pdays','previous']
```

```python
cleandata=X[myrows]
cleandata.head()
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
```

```python
cleandata.dtypes
```

```python
stringcolumns = cleandata.select_dtypes(include=object)
stringcolumns.columns
```

```python
cleandata.shape
```

```python
fig , axs = plt.subplots(int(len(stringcolumns.columns)/2),2, figsize=(10,12))
cleandata['job'].value_counts().plot(kind='bar',ax=axs[0,0])
cleandata['marital'].value_counts().plot(kind='bar',ax=axs[0,1])
cleandata['education'].value_counts().plot(kind='bar',ax=axs[1,0])
cleandata['default'].value_counts().plot(kind='bar',ax=axs[1,1])
cleandata['housing'].value_counts().plot(kind='bar',ax=axs[2,0])
cleandata['loan'].value_counts().plot(kind='bar',ax=axs[2,1])
fig.tight_layout()
fig.savefig('stringdatadistribution')
```
